### 练习2：温度转换

**任务**：输入摄氏度，输出华氏度和开尔文温度。

**公式**：
- 华氏度 = 摄氏度 × 9/5 + 32
- 开尔文 = 摄氏度 + 273.15

**提示**：
- 在Python中，乘号是`*`，除号是`/`
- 可以用`f-string`格式化输出，保留小数：`{value:.1f}`


````python
# 在这里写代码
celsius = 25  # 摄氏度

# 计算华氏度
fahrenheit = celsius * 9 / 5 + 32

# 计算开尔文
kelvin = celsius + 273.15

# 输出结果
print(f"{celsius}°C = {fahrenheit:.1f}°F")
print(f"{celsius}°C = {kelvin:.2f}K")
````

### 练习3：购物找零

**任务**：输入商品价格和付款金额，计算找零。

**要求**：
- 处理小数
- 输出格式：`应找零：XX.XX元`

**提示**：
- 用`float()`处理小数
- 用`f-string`格式化输出，保留2位小数：`{value:.2f}`


````python
# 在这里写代码
price = 45.8  # 商品价格
payment = 100  # 付款金额

# 计算找零
change = payment - price

# 输出结果
print(f"商品价格：{price}")
print(f"付款金额：{payment}")
print(f"应找零：{change:.2f}元")
````

### 练习4：秒数转换

**任务**：输入总秒数，转换成"X小时Y分钟Z秒"。

**提示**：
- 1小时 = 3600秒
- 1分钟 = 60秒
- 用`//`（整除）和`%`（取余）来计算

**示例**：
- 输入：3665秒
- 输出：1小时1分钟5秒


````python
# 在这里写代码
total_seconds = 3665  # 总秒数

# 计算小时
hours = total_seconds // 3600

# 计算剩余的秒数
remaining_seconds = total_seconds % 3600

# 计算分钟
minutes = remaining_seconds // 60

# 计算秒
seconds = remaining_seconds % 60

# 输出结果
print(f"{total_seconds}秒 = {hours}小时{minutes}分钟{seconds}秒")
````
